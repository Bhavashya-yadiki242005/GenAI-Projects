```python
#!pip install gradio pytesseract pillow langchain langchain_openai

import gradio as gr
from PIL import Image
import pytesseract
from langchain_openai import ChatOpenAI
from langchain.agents import initialize_agent, Tool
from langchain.prompts import PromptTemplate

# OCR function
def extract_text(image):
    return pytesseract.image_to_string(image)

# LangChain agent setup (Math Reasoning)
llm = ChatOpenAI(model="gpt-4", temperature=0)

math_prompt = PromptTemplate(
    input_variables=["question"],
    template="Solve this math related reasoning step by step and give only the final answer: {question}"
)

def solve_math(query):
    return llm.invoke(math_prompt.format(question=query)).content

# Pipeline: Image -> OCR -> Math Solution
def process_image(image):
    text = extract_text(image)
    if not text.strip():
        return "‚ö†Ô∏è No text detected in the image."
    answer = solve_math(text)
    return f"üìù Extracted Question:\n{text}\n\n‚úÖ Final Answer: {answer}"

# üé® Gradio UI
with gr.Blocks(css="footer {display:none !important;}") as demo:
    gr.Markdown(
        """
        # üßÆ Math Solver with LangChain and OCR
        Upload an image containing a math or logic question and this app will extract the text, understand it and solve step by step.
        """
    )

    with gr.Row():
        with gr.Column(scale=1):
            img_input = gr.Image(type="pil", label="üì§ Upload Your Question Image")
            clear_btn = gr.Button("üîÑ Clear")
        with gr.Column(scale=2):
            output_box = gr.Textbox(label="üìå Result", lines=10)

    img_input.upload(process_image, img_input, output_box)
    clear_btn.click(lambda: "", None, output_box)

demo.launch(share=True)
```
